```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import csv
#importing all the libraries
file_obj=open('startup_funding.csv',encoding='utf-8')
df=pd.read_csv(file_obj)
df=df["InvestorsName"]
#making a data frame
df=df.str.split(",",expand=True)
#As there are many investors present and we need to get each one as a value, so wew can use split method and seperate them in differnt columns
x=df.values
#making an array of differnt values of the investor , so we can get all values of data frame
a=[]
#making an empty list so that we can append the names.
for i in range(0,len(x)):
    for j in range(0,len(x[i])):
        if type(x[i][j]) == str: #removing none values
            a.append(x[i][j].strip()) #stripping  extra spaces
            
df= pd.DataFrame(a)
#converting into data frame so we can apply value counts method
val=df[0].value_counts()
x=val.index[0:5]
y=val.values[0:5]
for i in range (0,len(x)):
    print(x[i],y[i])  
plt.bar(x,y)
plt.xticks(rotation=45)
plt.xlabel("Investors")
plt.ylabel("Number of fundings")
plt.title("Investors vs number of fundings")
plt.show()
#plotting a bar graph of City and its relation with number of fundings
    
```
